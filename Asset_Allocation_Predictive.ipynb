```python
# Function to change Date column into pandas DateTime

def change_to_datetime(stocks_list):
  ''' Takes input list of assets '''

  df_list = []
  for i in range(len(stocks_list)):
    stocks_list[i]['Date'] = stocks_list[i]['Date'].apply(pd.to_datetime)
    df_list.append(stocks_list[i])
  return df_list
```

```python
# Function to get the specific range of data

def date_range(data, startrange, endrange):
  ''' Takes input data and start range and endrange,
   note that the ranges are of datetime type'''
  datalist = []
  for i in range(len(data)):
    datalist.append(data[i].set_index('Date').loc[startrange:endrange].reset_index())
  return datalist
```

# Forecast model


```python
# Get the data
y = []
for i in range(len(pv_list_df)):
  y.append(pv_list_df[i]['Close'].pct_change().fillna(0))
  
```

```python
# Scale the data
y_scaled = []
for i in range(len(y)): 
  scaler = MinMaxScaler(feature_range=(0,1))
  scaler = scaler.fit(y[i].to_numpy().reshape(-1, 1))
  y_scaled.append(scaler.transform(y[i].to_numpy().reshape(-1, 1)))
```

```python
# generate the input and output sequences

n_lookback = 60  # length of input sequences (lookback period)
n_forecast = 7  # length of output sequences (forecast period)

X = []
Y = []
for i in range(len(y_scaled)):
  X_temp = []
  Y_temp = []
  for j in range(n_lookback, len(y_scaled[i]) - n_forecast + 1):
    X_temp.append(y_scaled[i][j - n_lookback: j])
    Y_temp.append(y_scaled[i][j: j + n_forecast])
  X.append(X_temp)
  Y.append(Y_temp)
```

```python
X = [np.array(X[i]) for i in range(len(X))]
Y = [np.array(Y[i]) for i in range(len(Y))]
```

```python
len(X)
```

```python
# model
model = Sequential()
model.add(LSTM(units=50, return_sequences=True,
               input_shape=(n_lookback, 1)))
model.add(LSTM(units=50))
model.add(Dense(n_forecast))

model.compile(loss='mean_squared_error', optimizer='adam')
```

## Covariance

```python
# find correlation
corr = forecast_df.corr().to_numpy()
```

```python
corr[0,2]
```

```python
# Get the upper triangle elements
corr1 = corr[np.triu_indices(len(tickers), k=1)]
```

```python
corr1
```

```python
# Find the product of returns in order to calculate covariance i.e eg: ret of AAPL * ret of ANF and so on
mul = []
temp = gm_forecast_df.reset_index().drop('Date', axis=1)
for i in range(len(gm_forecast_df)):
  for j in range(len(gm_forecast_df)):
    if (j+i+1) <= len(gm_forecast_df):
      mul.append(temp.iloc[:,i].mean()*temp.iloc[:,j+i+1].mean())
```

```python
# Find the upper triangle part of covariance matrix
covarr = [corr1[i] * mul[i] for i in range(len(mul))]
```

```python
covarr
```

```python
# Replicate the upper triangle to lower as well 
cov_mat = forecast_df.corr()
cov_mat
```

```python
cov_mat1 = cov_mat.to_numpy()
# i_lower = np.tril_indices(n, -1)
cov_mat1[np.triu_indices(len(tickers), k=1)] = covarr
cov_mat1[np.tril_indices(len(tickers), -1)] = cov_mat1.T[np.tril_indices(len(tickers), -1)]
```

## Efficient Frontier 

```python
returns_daily = forecast_df
returns_annual = returns_daily.mean() * 250
```

```python
cov_daily = covariance_df
cov_annual = cov_daily * 250
```

```python
# empty lists to store returns, volatility and weights of imiginary portfolios
port_returns = []
port_volatility = []
sharpe_ratio = []
stock_weights = []
```

```python
# set the number of combinations for imaginary portfolios
num_assets = len(tickers)
num_portfolios = 50000
```

```python
#set random seed for reproduction's sake
np.random.seed(42)
```

```python
# populate the empty lists with each portfolios returns,risk and weights
for single_portfolio in range(num_portfolios):
    weights = np.random.random(num_assets)
    weights /= np.sum(weights)
    returns = np.dot(weights, returns_annual)
    volatility = np.sqrt(np.dot(weights.T, np.dot(cov_annual, weights)))
    sharpe = returns / volatility
    sharpe_ratio.append(sharpe)
    port_returns.append(returns)
    port_volatility.append(volatility)
    stock_weights.append(weights)
```

```python
# a dictionary for Returns and Risk values of each portfolio
portfolio = {'Returns': port_returns,
             'Volatility': port_volatility,
             'Sharpe Ratio': sharpe_ratio}
```

```python
# extend original dictionary to accomodate each ticker and weight in the portfolio
for counter,symbol in enumerate(tickers):
    portfolio[symbol+' Weight'] = [Weight[counter] for Weight in stock_weights]
```

```python
# make a nice dataframe of the extended dictionary
df = pd.DataFrame(portfolio)
```

```python
# plot frontier, max sharpe & min Volatility values with a scatterplot
plt.style.use('seaborn-dark')
df.plot.scatter(x='Volatility', y='Returns', c='Sharpe Ratio',
                cmap='RdYlGn', edgecolors='black', figsize=(10, 8), grid=True)
plt.xlabel('Volatility (Std. Deviation)')
plt.ylabel('Expected Returns')
plt.title('Efficient Frontier')
plt.show()
```

```python
# find min Volatility & max sharpe values in the dataframe (df)
min_volatility = df['Volatility'].min()
max_sharpe = df['Sharpe Ratio'].max()
```

```python
# use the min, max values to locate and create the two special portfolios
sharpe_portfolio = df.loc[df['Sharpe Ratio'] == max_sharpe]
min_variance_port = df.loc[df['Volatility'] == min_volatility]
```

```python
# plot frontier, max sharpe & min Volatility values with a scatterplot
plt.style.use('seaborn-dark')
df.plot.scatter(x='Volatility', y='Returns', c='Sharpe Ratio',
                cmap='RdYlGn', edgecolors='black', figsize=(10, 8), grid=True)
plt.scatter(x=sharpe_portfolio['Volatility'], y=sharpe_portfolio['Returns'], c='red', marker='D', s=200)
plt.scatter(x=min_variance_port['Volatility'], y=min_variance_port['Returns'], c='blue', marker='D', s=200 )
plt.xlabel('Volatility (Std. Deviation)')
plt.ylabel('Expected Returns')
plt.title('Efficient Frontier')
plt.show()
```

```python
# print the details of the 2 special portfolios
print('Minimum variance portfolio:\n',min_variance_port.T)
print('Maximum Sharpe ratio portfolio:\n',sharpe_portfolio.T)
```
